```python
import cv2
import os

from matplotlib import pyplot as plt
```

```python
IMG_PATH = os.path.join('/content/drive/MyDrive/Work/Practice/numpy/test_img', 'bike.jpg')
```

```python
img = cv2.imread(IMG_PATH)
img
```

```python
plt.imshow(img)
```

```python
#width * height
int(img.shape[1]), int(img.shape[0])
```

```python
#view image using openCV
from google.colab.patches import cv2_imshow #cv2/imshow() is not supported by google colab

resize_img = cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))
cv2_imshow(resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Canny edge detection algorithm

canny = cv2.Canny(img, threshold1=200, threshold2=210)
```

```python

resize_img = cv2.resize(canny, (int(img.shape[1]/2), int(img.shape[0]/2)))
cv2_imshow(canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#apply Gaussian blur to remove noise

blur = cv2.GaussianBlur(img, (5,5), 0)
```

```python
canny = cv2.Canny(blur, threshold1=200, threshold2=210)

cv2_imshow(canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('edges1.jpg', canny)
```

```python
#gray scaling

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray, (5,5), 0)
canny = cv2.Canny(blur, threshold1=200, threshold2=210)

cv2_imshow(canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('edges.jpg', canny)
```
